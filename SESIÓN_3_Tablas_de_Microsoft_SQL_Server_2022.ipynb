Una clave primaria (primary key en ingls) es un concepto fundamental en el diseo de bases de datos relacionales.

Una clave primaria es un atributo o conjunto de atributos que identifican de manera nica cada fila o registro en una tabla.

Al definir una clave primaria en una tabla, se garantiza que no puede haber duplicados ni valores nulos en los atributos que la componen.

Esto significa que cada fila de la tabla debe tener un valor nico en la clave primaria, lo que permite identificarla y acceder a ella de manera rpida y eficiente.



```
USE OTI_COMERCIAL
GO
IF EXISTS
(
	SELECT *
	FROM sys.schemas s
	INNER JOIN sys.tables t
	ON s.schema_id = t.schema_id
	WHERE
			s.name = 'Personal'
		AND
			t.name = 'Rol'
)
DROP TABLE Personal.Rol
GO
CREATE TABLE Personal.Rol
(
	cod_rol_in	INT PRIMARY KEY,
	nom_rol_vc	VARCHAR(20)
)
GO
INSERT INTO Personal.Rol(cod_rol_in,nom_rol_vc)
VALUES(5, 'Facturacin');
GO
INSERT INTO Personal.Rol(cod_rol_in,nom_rol_vc)
VALUES(2, 'Ventas');
GO
INSERT INTO Personal.Rol(cod_rol_in,nom_rol_vc)
VALUES(4, 'Compras');
GO
SET SHOWPLAN_ALL ON
GO
SELECT * FROM Personal.Rol
GO
SET SHOWPLAN_ALL OFF
GO
```

### **<font color="#FFFFFF">IDENTITY</font>**

</td>
</tr>
</table>

La propiedad IDENTITY se utiliza para generar automticamente valores numricos secuenciales para una columna en una tabla.

Es comnmente utilizado para asignar un identificador nico a cada fila de una tabla.

Cuando se define una columna con la propiedad IDENTITY, SQL Server se encarga de generar y asignar automticamente valores incrementales a esa columna cuando se inserta un nuevo registro en la tabla.

Los valores generados por la propiedad IDENTITY son nicos dentro de la tabla y no se repiten.



```
USE OTI_COMERCIAL
GO
IF EXISTS
(
	SELECT *
	FROM sys.schemas s
	INNER JOIN sys.tables t
	ON s.schema_id = t.schema_id
	WHERE
			s.name = 'Personal'
		AND
			t.name = 'Rol'
)
DROP TABLE Personal.Rol
GO
CREATE TABLE Personal.Rol
(
	cod_rol_in	INT IDENTITY(1,1)
	CONSTRAINT pk_personal_rol_cod_rol_in
	PRIMARY KEY CLUSTERED NOT NULL,
	nom_rol_vc	VARCHAR(20) NOT NULL
)
GO
INSERT INTO Personal.Rol(nom_rol_vc)
VALUES('Facturacin');
GO
INSERT INTO Personal.Rol(nom_rol_vc)
VALUES('Ventas');
GO
INSERT INTO Personal.Rol(nom_rol_vc)
VALUES('Compras');
GO
SELECT * FROM Personal.Rol
GO
```

```
USE OTI_COMERCIAL
GO
SELECT 'ALTER TABLE ['+s.name+'].['+t.name+'] '+
'ADD CONSTRAINT pk_'+s.name+'_'+
REPLACE(LOWER(t.name),' ','_')+
'_'+
REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(LOWER(c.name),' ','_'),'','a'),'','e'),'','i'),'','o'),'','u')+
' PRIMARY KEY CLUSTERED(['+ c.name + ']);'
FROM sys.schemas s
INNER JOIN sys.tables t
ON s.schema_id = t.schema_id
INNER JOIN sys.columns c
ON t.object_id = c.object_id
WHERE t.name NOT IN ('Rol', 'Ubigeo', 'sysdiagrams')
AND c.name LIKE 'ID%'
ORDER BY s.name, t.name, c.name
GO
```
