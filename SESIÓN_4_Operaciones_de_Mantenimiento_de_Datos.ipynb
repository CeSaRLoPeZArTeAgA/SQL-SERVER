La sentencia UPDATE en SQL Server 2022 se utiliza para modificar los registros existentes en una tabla de una base de datos.

Permite actualizar los valores de una o varias columnas en funcin de una condicin especificada.

</td>
</tr>

```
UPDATE esquema.nombre_tabla
SET columna1 = valor1, columna2 = valor2, ...
WHERE condicin;
```





```
USE OTI_COMERCIAL
GO
SELECT ID,dni_emp_ch, '0123456'+LTRIM(id)
FROM Personal.empleados
GO
```





```
USE OTI_COMERCIAL
GO
UPDATE Personal.empleados
SET dni_emp_ch = '0123456'+LTRIM(id)
GO
SELECT *
FROM Personal.empleados
GO
```





```
USE OTI_COMERCIAL
GO
UPDATE Personal.empleados
SET dni_emp_ch = '43535635'
WHERE ID= 2
GO
SELECT *
FROM Personal.empleados
GO
```





```
USE OTI_COMERCIAL
GO
UPDATE Personal.empleados
SET pwd_emp_vb = PWDENCRYPT('123')
WHERE ID=9
GO
SELECT *
FROM Personal.empleados
GO
```

## **<font color="#FFFFFF">DELETE</font>**

</td>
</tr>
</table>

La sentencia DELETE en SQL Server 2022 se utiliza para eliminar uno o varios registros de una tabla en una base de datos.

Permite eliminar datos de una tabla de acuerdo a una condicin especificada.


<table width='99%'>
<tr>
<td bgcolor='#81BC06'>

#### **<font color="#FFFFFF">Ejemplo</font>**

</td>
</tr>
</table>



```
USE master
GO
IF EXISTS
(
	SELECT name
	FROM sysdatabases
	WHERE name = 'prueba'
)
DROP DATABASE prueba
GO
CREATE DATABASE prueba
ON PRIMARY
(
	name = prueba_dat,
	filename = 'c:\db\prueba_dat.mdf',
	size = 30MB,
	maxsize = 50MB,
	filegrowth = 10MB
)
LOG ON
(
	name = prueba_log,
	filename = 'c:\db\prueba_log.ldf',
	size = 10MB,
	maxsize = UNLIMITED,
	filegrowth = 10%
)
GO
USE prueba
GO
CREATE TABLE amigo
(
	cod_ami_in INT IDENTITY(1,1)
	CONSTRAINT pk_amigo_cod_ami_in
	PRIMARY KEY CLUSTERED NOT NULL,
	nom_ami_vc	VARCHAR(35)
)
GO
INSERT INTO amigo (nom_ami_vc)
VALUES
('Carlos'),
('Sandra'),
('Julio'),
('Hugo')
GO
SELECT * FROM amigo
GO
DELETE FROM amigo
GO
SELECT * FROM amigo
GO
INSERT INTO amigo (nom_ami_vc)
VALUES
('Mara'),
('Silvia')
GO
SELECT * FROM amigo
GO
```

## **<font color="#FFFFFF">TRUNCATE TABLE</font>**

</td>
</tr>
</table>

La sentencia **TRUNCATE TABLE** en SQL Server 2022 se utiliza para eliminar todos los registros de una tabla de manera rpida y eficiente.

A diferencia de la sentencia DELETE, que elimina los registros uno por uno, TRUNCATE TABLE elimina todos los datos de la tabla de una sola vez.

Esta operacin es irreversible y no se puede deshacer.

<table width='99%'>
<tr>
<td bgcolor='#81BC06'>

#### **<font color="#FFFFFF">Ejemplo</font>**

</td>
</tr>
</table>



```
USE master
GO
IF EXISTS
(
    SELECT name
    FROM sysdatabases
    WHERE name = 'prueba'
)
DROP DATABASE prueba
GO
CREATE DATABASE prueba
ON PRIMARY
(
    name = prueba_dat,
    filename = 'c:\db\prueba_dat.mdf',
    size = 30MB,
    maxsize = 50MB,
    filegrowth = 10MB
)
LOG ON
(
    name = prueba_log,
    filename = 'c:\db\prueba_log.ldf',
    size = 10MB,
    maxsize = UNLIMITED,
    filegrowth = 10%
)
GO
USE prueba
GO
CREATE TABLE amigo
(
    cod_ami_in INT IDENTITY(1,1)
    CONSTRAINT pk_amigo_cod_ami_in
    PRIMARY KEY CLUSTERED NOT NULL,
    nom_ami_vc    VARCHAR(35)
)
GO
INSERT INTO amigo (nom_ami_vc)
VALUES
('Carlos'),
('Sandra'),
('Julio'),
('Hugo')
GO
SELECT * FROM amigo
GO
TRUNCATE TABLE amigo
GO
SELECT * FROM amigo
GO
INSERT INTO amigo (nom_ami_vc)
VALUES
('Mara'),
('Silvia')
GO
SELECT * FROM amigo
GO
```
