| Operador | Descripción                                    |
|----------|------------------------------------------------|
| =        | Igual a                                        |
| <> o !=       | Diferente de                                   |
| >        | Mayor que                                      |
| <        | Menor que                                      |
| >=       | Mayor o igual que                              |
| <=       | Menor o igual que                              |
| BETWEEN  | Está entre dos valores                         |
| LIKE     | Coincide con un patrón                         |
| IN       | Está dentro de un conjunto de valores           |
| NOT      | Niega una condición                            |
| AND      | Cumple ambas condiciones                        |
| OR       | Cumple al menos una de las condiciones          |
| NOT IN   | No está dentro de un conjunto de valores        |
| NOT LIKE | No coincide con un patrón                       |
| IS NULL  | Es nulo (no tiene valor)                        |
| IS NOT NULL | No es nulo (tiene valor)                       |

El carácter % permite remplazar a un conjunto de carácteres.



```
/*
Se desea visualizar el listado de beneficiarios
cuyo apellido paterno comience con la letra w
*/
USE pvl
GO
SELECT
	pat_ben_vc AS PATERNO,
	mat_ben_vc AS MATERNO,
	nom_ben_vc AS NOMBRES
FROM Beneficiario
WHERE pat_ben_vc LIKE 'w%'
ORDER BY 1, 2, 3
```





```
/*
Se desea visualizar el listado de beneficiarios
cuyo apellido paterno termine con la letra y
*/
USE pvl
GO
SELECT
	pat_ben_vc AS PATERNO,
	mat_ben_vc AS MATERNO,
	nom_ben_vc AS NOMBRES
FROM Beneficiario
WHERE pat_ben_vc LIKE '%y'
ORDER BY 1, 2, 3
```





```
/*
Se desea visualizar el listado de beneficiarios
cuyo apellido paterno contenga la palabra AMOR
*/
USE pvl
GO
SELECT
	pat_ben_vc AS PATERNO,
	mat_ben_vc AS MATERNO,
	nom_ben_vc AS NOMBRES
FROM Beneficiario
WHERE pat_ben_vc LIKE '%AMOR%'
ORDER BY 1, 2, 3
```

El carácter de subrayado "_" permite remplazar a un solo carácter.



```
/*
Se desea visualizar el listado de beneficiarios
cuyo apellido paterno contenga en su tercer carácter
la letra S
*/
USE pvl
GO
SELECT
	SUBSTRING(pat_ben_vc, 3, 1) AS [Tercer carácter],
	pat_ben_vc AS PATERNO,
	mat_ben_vc AS MATERNO,
	nom_ben_vc AS NOMBRES
FROM Beneficiario
WHERE pat_ben_vc LIKE '__S%'
ORDER BY 1, 2, 3, 4
```
