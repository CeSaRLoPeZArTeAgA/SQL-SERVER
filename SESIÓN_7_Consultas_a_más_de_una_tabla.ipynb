## **<font color="#000000">ROW_NUMBER()</font>**

</td>
</tr>
</table>

 La funcin ROW_NUMBER() se utiliza en la clusula SELECT para asignar un nmero secuencial a cada fila resultante de una consulta.

 Esta funcin se usa comnmente para generar nmeros de fila que pueden ser tiles en diferentes escenarios, como la paginacin de resultados o la identificacin de filas duplicadas.



```
USE OTI_COMERCIAL
GO
SELECT
	ROW_NUMBER() OVER (PARTITION BY Categora ORDER BY Categora) AS Item,
	Id, [Nombre del producto], Categora
FROM Almacen.Productos
ORDER BY Categora, [Nombre del producto]
```

##### **<font color="#FFFFFF">Clculo del Stock de los productos</font>**

</td>
</tr>
</table>



```
USE OTI_COMERCIAL
GO
SELECT Id, [Cdigo de producto], [Nombre del producto],
SUM
(
	CONVERT
	(
		INT,
		CASE t.[Tipo de transaccin]
			WHEN 1 THEN '+'
			WHEN 2 THEN '-'
			WHEN 3 THEN '-'
			WHEN 4 THEN '+'
		END + LTRIM(t.Cantidad)
	)
) AS Stock
FROM Almacen.Productos p
INNER JOIN Almacen.[Transacciones de inventario] t
ON p.Id=t.[Id de producto]
GROUP BY Id, [Cdigo de producto], [Nombre del producto]
ORDER BY 4 DESC
```

## **<font color="#000000">NULLIF()</font>**

</td>
</tr>
</table>

La funcin NULLIF se utiliza para comparar dos expresiones y devolver NULL si ambas son iguales.

Si las dos expresiones son diferentes , se devuelve el valor de la primera expresin.




```
USE Northwind
GO
SELECT
	OrderID, OrderDate,
	CONVERT(CHAR(10), OrderDate, 103) AS Fecha1,
	NULLIF(CONVERT(CHAR(10), OrderDate, 103), '01/04/1998') AS Fecha2
FROM Orders
```

## **<font color="#000000">GROUPING SET</font>**

</td>
</tr>
</table>

La clusula GROUPING SETS se utiliza en una clusula GROUP BY para especificar mltiples conjuntos de agrupacin en una sola consulta.

Esta clusula permite realizar agregaciones y generar resultados resumidos para diferentes niveles de agrupacin en una nica consulta.



```
USE Northwind
GO
SELECT
	p.ProductName AS Producto,
	YEAR(o.orderdate) AS Ao,
	DATENAME(MONTH, o.OrderDate) AS Mes,
	CONVERT(DECIMAL(19,2),SUM(od.Quantity* od.UnitPrice * (1-od.Discount))) AS Monto
FROM Products p
INNER JOIN [Order Details] od
ON p.ProductID = od.ProductID
INNER JOIN orders o
ON od.orderid = o.orderid
GROUP BY GROUPING SETS
(
	p.ProductName,
	YEAR(o.orderdate),
	DATENAME(MONTH, o.OrderDate)
)
```

## **<font color="#000000">ROLLUP</font>**

</td>
</tr>
</table>

La clusula ROLLUP se utiliza en una clusula GROUP BY para generar resultados resumidos en diferentes niveles de agrupacin jerrquica.

Esta clusula permite realizar agregaciones y obtener totales y subtotales en una sola consulta.



```
USE Northwind
GO
SELECT
    c.CategoryName AS Categora,
	o.ShipCountry AS Pas,
	CONVERT(DECIMAL(19,2),SUM(od.Quantity* od.UnitPrice * (1-od.Discount))) AS Monto
FROM Categories c
INNER JOIN Products p
ON c.CategoryID= p.CategoryID
INNER JOIN [Order Details] od
ON p.ProductID = od.ProductID
INNER JOIN orders o
ON od.orderid = o.orderid
GROUP BY
    c.CategoryName,
	o.ShipCountry WITH ROLLUP

```

## **<font color="#000000">CUBE</font>**

</td>
</tr>
</table>

La clusula **CUBE** se utiliza junto con la funcin GROUP BY para generar resultados de agregacin multidimensionales.

Al igual que la clusula ROLLUP, la clusula CUBE permite calcular subtotales y totales generales, pero a diferencia de ROLLUP, CUBE genera todas las combinaciones posibles de los valores de las columnas especificadas.



```
USE Northwind
GO
SELECT
    c.CategoryName AS Categora,
	o.ShipCountry AS Pas,
	CONVERT(DECIMAL(19,2),SUM(od.Quantity* od.UnitPrice * (1-od.Discount))) AS Monto
FROM Categories c
INNER JOIN Products p
ON c.CategoryID= p.CategoryID
INNER JOIN [Order Details] od
ON p.ProductID = od.ProductID
INNER JOIN orders o
ON od.orderid = o.orderid
GROUP BY
    c.CategoryName,
	o.ShipCountry WITH CUBE
```
