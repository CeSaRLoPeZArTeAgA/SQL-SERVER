



```
USE master
GO
IF EXISTS
(
	SELECT name
	FROM sysdatabases
	WHERE name = 'Colegio'
)
DROP DATABASE Colegio
GO
CREATE DATABASE Colegio
ON PRIMARY
(
	name = 'colegio_dat',
	filename = 'c:\db\colegio_dat.mdf',
	size=30MB,
	maxsize=30MB,
	filegrowth=10MB
)
LOG ON
(
	name = 'colegio_log',
	filename = 'c:\db\colegio_log.ldf',
	size=10MB,
	maxsize=UNLIMITED,
	filegrowth=10%
)
GO
```

En SQL Server 2022, un esquema es un contenedor lógico que se utiliza para organizar y agrupar objetos de base de datos, como tablas, vistas, procedimientos almacenados, funciones y otros objetos relacionados. Proporciona una forma de estructurar y separar los objetos dentro de una base de datos.

Un esquema actúa como un espacio de nombres dentro de una base de datos y permite evitar conflictos de nombres entre objetos. Cada objeto en una base de datos debe pertenecer a un esquema.



```
USE AdventureWorks2022
GO
SELECT name
FROM sys.schemas
WHERE principal_id<>schema_id
ORDER BY name
GO
```





```
USE OTI_COMERCIAL
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Ventas'
)
DROP SCHEMA Ventas
GO
CREATE SCHEMA Ventas
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Almacen'
)
DROP SCHEMA Almacen
GO
CREATE SCHEMA Almacen
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Facturacion'
)
DROP SCHEMA Facturacion
GO
CREATE SCHEMA Facturacion
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Cobranzas'
)
DROP SCHEMA Cobranzas
GO
CREATE SCHEMA Cobranzas
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Compras'
)
DROP SCHEMA Compras
GO
CREATE SCHEMA Compras
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Pagos'
)
DROP SCHEMA Pagos
GO
CREATE SCHEMA Pagos
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Personal'
)
DROP SCHEMA Personal
GO
CREATE SCHEMA Personal
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Sistemas'
)
DROP SCHEMA Sistemas
GO
CREATE SCHEMA Sistemas
GO
IF EXISTS
(
	SELECT name
	FROM sys.schemas
	WHERE name = 'Transportes'
)
DROP SCHEMA Transportes
GO
CREATE SCHEMA Transportes
GO
SELECT name
FROM sys.schemas
WHERE principal_id<>schema_id
ORDER BY name
GO
```
